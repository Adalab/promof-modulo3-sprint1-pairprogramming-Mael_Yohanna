Elegi merge inner porque hay una columna que coincide, que es "Country", y queremos todos los datos de las dos tablas.

```python
df_1 = pd.read_csv("world-data-2023_part1.csv", index_col=0)
df_1.head()
```

```python
df_2 = pd.read_csv("world-data-2023_part2.csv", index_col=0)
df_2.head()
```

```python
df_1.columns
```

```python
df_2.columns
```

```python
merge_inner = df_1.merge(df_2, left_on='Country', right_on = "country")
merge_inner.head()
```

1. Después de la unión de datos, tenemos dos columnas de "country". Elimina una de ellas.

```python
merge_inner.drop("country", axis = 1, inplace = True)
merge_inner.head(2)
```

2. Los nombres de las columnas no son homogeneos. Cambia los nombres de las columnas de tal forma que:
- No tengan espacios.
- Estén en minúscula.
- No tengan paréntesis, es decir, quitar "(%)", "(Km2)".
- Algunas columnas tiene "\n". Eliminalos de los nombres de las columnas.
- Algunas columnas tienen ":". Eliminalos de los nombres de las columnas.

```python
df_final = merge_inner.copy()
df_final.head()
```

```python
nuevas_columnas = {columna : columna.lower().replace(" ", "").replace(':', '').split("(")[0].strip() for columna in df_final.columns}
nuevas_columnas
```

```python
df_final.rename(columns = nuevas_columnas, inplace = True)

# comprobamos que se han cambiado los nombres de las columnas
df_final.head()
```

3. La columnas coordinates tiene la latitud y la longitud en una sola columna. Crea dos columnas nuevas, una con la longitud y otra con la latitud. Una vez hecho, elimina la columna de coordinates.

```python
df_final[["lat","long"]] = df_final["coordinates"].str.split(",", expand = True)
df_final.drop("coordinates", axis = 1, inplace = True)
df_final.head(2)
```

```python
df_final['long'] = df_final['long'].str.replace(r"[()'']", "", regex=True)
df_final['lat'] = df_final['lat'].str.replace(r"[()'']", "", regex=True)
df_final.head(5)
```

4. Las columnas unemployment_rate, total_tax_rate, tax_revenue, population_labor_force_participation, out_of_pocket_health_expenditure, gross_tertiary_education_enrollment, gross_primary_education_enrollment, forested_area, cpi_change, agricultural_land tienen "%". Elimina los "%" de los valores de las columnas.

```python
columnas = [
    "unemploymentrate", 
    "totaltaxrate", 
    "taxrevenue", 
    "populationlaborforceparticipation", 
    "outofpockethealthexpenditure", 
    "grosstertiaryeducationenrollment", 
    "grossprimaryeducationenrollment", 
    "forestedarea", 
    "cpichange", 
    "agriculturalland"
]

for column in columnas:
    df_final[column] = df_final[column].str.rstrip('%')

df_final.head(5)
```

5. Haz lo mismo para las columnas de gasoline_price, gdp, minimum_wage, pero eliminando "$".

```python
columnas_1 = [
    "gasolineprice", 
    "gdp", 
    "minimumwage", 
   
]

for column in columnas_1:
    df_final[column] = df_final[column].str.lstrip('$')

df_final.head(5)
```
