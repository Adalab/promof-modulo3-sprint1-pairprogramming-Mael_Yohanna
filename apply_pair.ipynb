Las columnas land_area, armed_forces_size, co2-emissions, gdp, population, urban_population, latitude, longitude, density, cpi son de tipo object pero en realidad deberían ser numéricas. Crea una función que nos permita cambiar el tipo de dato de estas columnas al que le corresponde.

```python
df_1.info()
```

```python
def cambiar_comas(cadena):
     try:
          return float(cadena.replace("'","").replace(',',''))
     except:
          return np.nan
```

```python
lista_columnas = ["landarea", "armedforcessize", "co2-emissions", "gdp", "population", "urban_population", "lat", "long", "density", "cpi"]
```

```python
for col in lista_columnas:
    df_1[col] = df_1[col].apply(cambiar_comas)
df_1
```

2. Crea una función para añadir una nueva columna llamada 'Continent' al DataFrame para indicar a qué continente pertenece cada país.

```python
df_1.rename(columns={'long':'lat','lat':'long'},inplace=True)
```

```python
def continent (lat,long):
    if -55 <= lat <= 90 and -170 <= long <=- 25:
        return 'America'
    elif 35 <= lat <= 70 and -10 <= long <= 40:
        return 'Europa'
    elif -10 <= lat <= 80 and 30 <= long <= 180:
        return 'Asia'
    elif -35 <= lat <= 35 and -20 <= long <= 50:
        return 'Africa'
    else:
        return 'Oceania'
    
```

```python
df_1['continent']=df_1.apply(lambda dataframe: continent (dataframe['lat'],dataframe['long']), axis=1)
```

5. ¿Cuál es la tasa de impuestos total máxima por densidad de población por continente? Para solucionar este ejercicio puedes:

- Primero, necesitas calcular la tasa de impuestos total para cada continente. La tasa de impuestos total se encuentra en la columna "total_tax_rate".
- Luego, calcula la densidad de población para cada continente. La densidad de población se encuentra en la columna "density".
- Ahora, divide la tasa de impuestos total entre la densidad de población para obtener la tasa de impuestos total por persona por kilómetro cuadrado.

```python
df_1.columns
```

```python
tasa_imput = df_1.groupby("continent")["totaltaxrate"].sum().reset_index().round(2)
tasa_imput
```

```python
densidad_tasa = tasa_imput.merge(densidad, how = "inner", on = "continent")
densidad_tasa
```

```python
def tasa_por_persona(tasa, density):
    return tasa/density
```

```python
densidad_tasa["tasa_por_persona"] = densidad_tasa.apply(lambda df: tasa_por_persona(df["totaltaxrate"], df['density']),axis=1)
```

```python
densidad_tasa[["continent","tasa_por_persona"]].max()
```
